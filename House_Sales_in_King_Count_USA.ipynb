

Fit a linear regression model to predict the price using the list of features:


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]   
#finding the r^2 score for each individual features  
for feature in features:
    x = df[[feature]]
    y = df[['price']]
    lm = LinearRegression()
    lm.fit(x,y)
    print('The R-squared value of the model using only',feature, 'as feature is: ', lm.score(x,y))
```

Then calculate the R^2 for the combined features.


```python
#find the r^2 score with all the listed features combined
x = df[features]
y = df[['price']]
lm = LinearRegression()
lm.fit(x,y)
score = lm.score(x,y)
print('The R^2 score of the model using the combined features is ', score)
```

Creating a pipeline object to predict the 'price', I will fit the object using the features in the list 'features', and calculate the R^2.


```python
#first create a list of input
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

```python
#performing the pipeline activity on each feature
for feature in features:
    x = df[[feature]]
    y = df[['price']]
    pipeline = Pipeline(Input)
    pipeline.fit(x,y)
    score = pipeline.score(x,y)
    print('The R-squared value of the model using only',feature, 'as feature is: ', score)
```

```python
#now to perform the pipeline with the list of features combined
x = df[features]
y = df[['price']]
pipeline = Pipeline(Input)
pipeline.fit(x,y)
score = pipeline.score(x,y)
print('The R-squared value of the model using all the features is: ', score)
```

# Task 5: Model Evaluation and Refinement


Import the necessary modules:


```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
```

I will split the data into training and testing sets:


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```



Create and fit a Ridge regression object using the training data, set the regularization parameter to 0.1, and calculate the R^2 using the test data.


```python
from sklearn.linear_model import Ridge
```

```python
RR  = Ridge(alpha=0.1)
RR.fit(x_train, y_train)
score = RR.score(x_train, y_train)
print('The R^2 score is ', score)
```

### Practicing the polynomial transformation of training data

Perform a second order polynomial transform on both the training data and testing data. Create and fit a Ridge regression object using the training data, set the regularisation parameter to 0.1, and calculate the R^2 utilising the test data provided. Take a screenshot of your code and the R^2.


```python
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
RidgeModel = Ridge(alpha=0.1)
RidgeModel.fit(x_train_pr, y_train)
print('The R-square is: ', RidgeModel.score(x_test_pr, y_test))
```
